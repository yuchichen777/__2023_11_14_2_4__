```python
import pandas as pd

dataFrame = pd.read_excel('year2021.xlsx',usecols=['國家','日期','總確診數','新增確診數'])
dataFrame
```

```python
dataFrame.query('新增確診數>10000')
```

```python
#有了排序,可以找最大前幾筆,最小前幾筆
sortedFrame = dataFrame.sort_values(by='新增確診數',ascending=False)
sortedFrame.iloc[:20]
```

```python
from sqlalchemy import create_engine
engine = create_engine("sqlite:///year2021.db")

dataFrame.to_sql('台灣日本2021',engine,if_exists='replace',index=False)
```
