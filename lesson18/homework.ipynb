```python
#儲存為cov19.db,資料表名稱:台灣2021
import pandas as pd
from sqlalchemy import create_engine

dataFrame = pd.read_csv('world.csv')
dataFrame = dataFrame.query('國家=="台灣" and 日期>="2021-01-01" and 日期<="2021-12-31"')

engine = create_engine("sqlite:///cov19.db")

dataFrame.to_sql('台灣2021',engine,if_exists='replace',index=False)
```

```python
#1. 讀取world.csv
#2. 選取美國,新增確診數最多的前100筆
#3. 儲存為json檔
import pandas as pd

dataFrame = pd.read_csv('world.csv')
dataFrame = dataFrame.query('國家=="美國"')

sortedFrame = dataFrame.sort_values(by='新增確診數', ascending=False)
sortedFrame.iloc[:100].to_json('美國top100新增確診數.json', orient='table', force_ascii='False')
```
