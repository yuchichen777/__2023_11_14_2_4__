```python
import pandas as pd

dataFrame = pd.read_csv("world.csv")
dataFrame.info()
```

```python
allDataFrame = dataFrame.reindex(columns=['洲名','國家','日期','總確診數','新增確診數','總人口數','解封指數'])
allDataFrame
```

```python
allDataFrame.info()
allDataFrame['日期'] = pd.to_datetime(allDataFrame['日期'])
allDataFrame.info()
```

```python
display(allDataFrame)
#使用subscript
allDataFrame[allDataFrame['洲名'] == '亞洲']
```

```python
display(allDataFrame.query('國家=="台灣"'))
display(allDataFrame.query('國家=="日本"'))
```

```python
taiwan_japan = allDataFrame.query('國家=="台灣" or 國家=="日本"')
taiwan_japan
```

```python
#datatime的搜尋
#取出2020-01-22
taiwan_japan.query('日期=="2020-01-22"')
```

```python
#取出小於2020-02-01
taiwan_japan.query('日期<"2020-02-01"')
```

```python
year2021 = taiwan_japan.query('日期>="2021-01-01" and 日期<"2022-01-01"')
year2021
```

```python
year2021.to_excel('year2021.xlsx')
```
