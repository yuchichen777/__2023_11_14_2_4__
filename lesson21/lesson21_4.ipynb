```python
import numpy as np
import pandas as pd

df = pd.read_csv('各鄉鎮市區人口密度.csv')
df.columns = df.iloc[0]
df1 = df.drop(index=0)
```

```python
df2 = df1.dropna()
df2
```

```python
df2.iloc[369]
df3 = df2.replace(to_replace="… ",value=0)
df3
```

```python
df3[['年底人口數','人口密度']] = df3[['年底人口數','人口密度']].astype(int)
df3['土地面積'] = df3['土地面積'].astype(float)
df3.info()
```

```python
def split(v:str) -> str:
    if len(v)>4:
        return v[:3]
    else:
        return v
    
df3['區域別'] = df3['區域別'].map(split)
```
