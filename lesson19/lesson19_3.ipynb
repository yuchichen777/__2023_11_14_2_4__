```python
import pandas as pd

allData = pd.read_csv('world.csv',
                      usecols=['國家', '日期', '總確診數', '新增確診數'],
                      index_col=['日期'])

allData
```

```python
allData.index = pd.to_datetime(allData.index)
allData.info()
```

```python
allData.loc['2021-05']
```

```python
taiwan = allData.query('國家=="台灣"')
japan = allData.query('國家=="日本"')
```

```python
taiwan1 = taiwan[['總確診數','新增確診數']]
taiwan1.columns = ['台灣總確診數','台灣新增確診數']
taiwan1
```

```python
japan1 = japan[['總確診數','新增確診數']]
japan1.columns = ['日本總確診數','日本新增確診數']
japan1
```

```python
allDataFrame = pd.concat([taiwan1,japan1],axis=1)
newDataFrame = allDataFrame.dropna()
newDataFrame
```

```python
def callBack(value:pd.Series):
    return pd.Series([value.min(),value.max(),value.median(),value.mean(),value.std()],
                     index=['最少','最多','中間數','平均','標準差'])

newDataFrame[['台灣新增確診數','日本新增確診數']].apply(callBack)
```
